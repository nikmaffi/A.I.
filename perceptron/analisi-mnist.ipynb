```python
from sklearn.model_selection import train_test_split
import lib.neuron as neuron
import seaborn as sns
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_csv("./data/mnist.csv")
data
```

```python
#Rimozione prima colonna (indici delle righe)

data.pop("garbage")
data = data[data["target"].isin([0, 1])]
data
```

```python
X = data.copy()
y = X.pop("target")
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 0)
```

```python
n = neuron.Neuron(784, 0.17, lambda x: 1 / (1 + np.exp(-x)), lambda x: (1 / (1 + math.exp(-x))) * (1 - (1 / (1 + math.exp(-x)))))
n.fit(X_train.values, y_train.values, 100)
```

```python
print(n.score(X_test.values, y_test.values))
```
