```python
num_data = data[data["emotion"].isin([0, 2])].copy()
new_data = {}

for i, string in enumerate(num_data["pixels"].values):
    pixels = np.array(string.split(" "), dtype=int)

    if i == 0:
        new_data["emotion"] = []
        for k in range(pixels.size):
            new_data[str(k)] = []

    new_data["emotion"].append(num_data["emotion"].values[i])

    for j, pixel in enumerate(pixels):
        new_data[str(j)].append(pixel)

num_data = pd.DataFrame(new_data)
num_data
```

```python
X_not_norm = num_data.loc[:, num_data.columns != "emotion"].copy()
X_not_norm
```

```python
y = num_data["emotion"].copy()
y
```

```python
m = X_not_norm.mean()
s = X_not_norm.std()

X = (X_not_norm - m) / s
X
```

```python
training_set_X, test_set_X, training_set_y, test_set_y = train_test_split(X, y, train_size = 0.7, random_state = 0)
```

```python
n = neuron.Neuron(len(X.columns), 0.17, lambda x: 2 if np.arctan(x) > 0 else 0, lambda x: 1 / (1 + x ** 2))
n.fit(training_set_X.values, training_set_y.values, 100)
```

```python
input_data = test_set_X.values
expeted_output_data = test_set_y.values

print(n.score(input_data, expeted_output_data))
```
